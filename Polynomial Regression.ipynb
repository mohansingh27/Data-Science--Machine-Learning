```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sal= pd.read_csv('/content/Position_Salaries.csv')
```

```python
df_sal.head()
```

```python
df_sal.tail()
```

```python
df_sal.describe()
```

```python
sns.distplot(df_sal['Salary'])
plt.title('Salary Distribution')
plt.show
```

```python
# Scatter plot

plt.scatter(df_sal['Level'],df_sal['Salary'])
plt.title('Salary vs Level')
plt.xlabel('Levels')
plt.ylabel('Salary')
plt.show()
```

```python
# Splitting the data

x= df_sal.iloc[:,1:2].values # Independent(Levels)
y=df_sal.iloc[:,2].values # Dependent(Salary)
print(x)
print(y)
```

```python
# Train Linear Regression Model
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x,y)
```

```python
# Predicting results

y_pred_lr= lr.predict(x)
```

```python
# Visualize LR Data
plt.scatter(x,y,color='red')
plt.plot(x,lr.predict(x),color='Blue')
plt.title('Linear regression')
plt.xlabel('Levels')
plt.ylabel('Salary')
plt.show()
```

```python
# Ploynomial Regression

from  sklearn.preprocessing import PolynomialFeatures

pr= PolynomialFeatures(degree=4)
x_poly= pr.fit_transform(x)
lr2= LinearRegression()
lr2.fit(x_poly,y)
```

```python
# Predicting results

y_pred_poly= lr2.predict(x_poly)
```

```python
# Visualiztion

plt.scatter(x,y,color='red')
plt.plot(x,lr2.predict(x_poly),color='green')
plt.xlabel('Levels')
plt.ylabel('Salries')
plt.title('Polynomial Regression')
plt.show()
```

```python
# Testing the acuracies

lr.predict([[6]]) # Linear Regression
```

```python
lr2.predict(pr.fit_transform([[6]])) # Polynomial Regression
```
