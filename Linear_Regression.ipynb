```python
import pandas as pd # Data retrieval
import numpy as np # Mathematical functions
import matplotlib.pyplot as plt # Plotting graph
import seaborn as sns # Visualiztion
from sklearn.model_selection import train_test_split # Splitting the data as train and test set.
from pandas.core.common import random_state # To define random state
from sklearn.linear_model import LinearRegression # importing linear regression model
```

```python
sal_data= pd.read_csv('/content/Salary_Data.csv')
```

```python
# Data Distribution
sns.displot(sal_data['Salary'])
plt.title('salary distribution data')
plt.show()
```

```python
# Corelation between experience and salry.
plt.scatter(sal_data['YearsExperience'], sal_data['Salary'],color= 'red')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(True)
plt.show()
```

```python
# Splitting variables
X = sal_data.iloc[:, :1] # independent
y = sal_data.iloc[:, 1:] # dependent
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 0)
```

```python
# Regressor model
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Prediction result
y_test = lr.predict(X_test) # predicted value of y_test
y_train = lr.predict(X_train) # predicted value of y_train
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, y_train, color = 'green')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(True)
plt.show()
```

```python
# Prediction on testing set
plt.scatter(X_test, y_test, color = 'blue')
plt.plot(X_test, y_test, color = 'green')
plt.title('Salary vs Experience (Testing Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(True)
plt.show()
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {lr.coef_}')
print(f'Intercept: {lr.intercept_}')
```

```python
y_pred=lr.predict([[1.5],[2]])
print(y_pred.round())
```

```python
# Calculating Mean Absolute Error(MAE)

from sklearn.metrics import mean_absolute_error
mae= mean_absolute_error(y_test, y_pred)
print("Mean absolute error:", mae)
```

```python
from sklearn.metrics import mean_squared_error
mse= mean_absolute_error(y_test, y_pred)
print("Mean squared error:", mse)
```

```python
from sklearn.metrics import mean_squared_error
rmse= mean_absolute_error(y_test, y_pred)
print("Root Mean Squared Error:", rmse)
```
